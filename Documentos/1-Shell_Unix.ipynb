Introducción a la shell Unix
======================

1.- El software libre y los sistemas operativos Unix
----------------------

Aunque muchos usuarios lo desconozcan, la mayor parte del software que mantiene internet, así como compiladores y sitemas para computación científica, tienen **licencia libre** \[[1](https://directory.fsf.org/), [2](https://en.wikipedia.org/wiki/Category:Free_science_software)\]. Esto significa que sus desarolladores han optado por mantener abierto el código de estos programas, permitiendo su reutilización por terceros y, a su vez, aprovechando las sinergias de la cloaboración, la investigación y el aprendizaje del trabajo del otras personas. 

Esta es, de hecho, la forma de habitual de relacionarse en comunidades científicas. Aplicada al campo del desarrollo de software, genera oportunidades de negocio distintas a las aquellas otros modelos de desarrollo de software basados en la privación de derechos (de estudio, de copia, de distribución, de ejecución) a los usuarios [\[3\]](https://www.gnu.org/philosophy/fsfs/free_software.es.pdf).

A partir del nacimiento de la *free software foundation* en los años 80 se desarrolló un sistema operativo, conocido como [GNU/Linux](https://www.gnu.org/gnu/linux-and-gnu.es.html), completamente libre. Este sitema operativo es de tipo Unix (y por tanto muy potente, multitarea, multiusuario, etc) y es el más utilizado entre los [ordenadores más potentes del mundo](https://www.top500.org/), que están orientados a la computación de alto rendimiento. 

Y a su vez, Linux es la base de decenas de *miles de millones* de pequeños y omnipresentes ordenadores (*smartphones*). En efecto [Android está basado en el núcleo Linux](https://es.wikipedia.org/wiki/Android) (imagen tomada de Wikimedia Commons). <img src="https://upload.wikimedia.org/wikipedia/commons/8/83/World_Wide_Smartphone_Sales.png" alt="Comparativa" style="width: 70%"/>.

2.- La *shell* Unix
---------------------

Para utilizar un gran servidor de cálculo paralelo, se emplean interfaces basadas en líneas de órdenes, conocidas como *shells Unix*. estas interfaces también tienen sus ventajas en ordenadores personales. Por ejemplo, para el tratamiento de grandes volúmnes de datos y, en general, para todos los procesos relacionados con la ciencia de datos o *big data*. Aunque más compleja que las usuales interfaces de usuario basadas en el ratón, su dominio posibilita la automatización y la realización de tareas que de otra forma no serían factibles.

Una *shell* es un programa que te permite interactuar con el sistema operativo del ordenador (y, a través de él, con el software y el hardware de tu equipo) escribiendo órdenes o *comandos*. El intérprete de la shell interpreta la orden y, típicamente, realiza alguna acción como resultado. Algunas veces a la shell se la conoce con las siglas CLI (del ingés *comand line interface*), como opuesto a la *interfaz gráfica de usuario* o GUI (del inglés
*graphical user interface*), que consiste de señalar y hacer "click". La shell Unix más extendida es la denominada [BASH](https://www.gnu.org/software/bash/manual/html_node/index.html).


Hay muchos comandos en una shell Unix y la mayoría de ellos tienen muchos
argumentos opcionales (generalmente especificados añadiendo algo como -x
después del nombre del comando, donde x es una letra). Solo algunos
comandos importantes serán revisados aquí. Ver por
ejemplo, [Wikipedia-unix-utilities](https://en.wikipedia.org/wiki/List_of_Unix_commands) para links con más detalles.

En un sistem **Windows**, la mayoría de la gente su entorno gráfico (su GUI) usual, aunque también es
posible abrir una ventana en un modo línea de comando para su sistema operativo DOS. Nota que DOS es
diferente de Unix. Otra posibilidad es usar [Cygwin](https://www.cygwin.com/), un programa que ejecuta una shell Unix dentro de Windows. 

A partir de Windows 10 es posible, incluso, ejecutar un escritorio de tipo GNU/Linux (en concreto, de la distribución [Ubuntu](https://www.ubuntu.com/)). Aunque más recomendable es instalar un sistema opeativo GNU/Linux dentro de una máquina virtual (para ello, por ejemplo, se puede usar [VirtualBox](https://www.virtualbox.org/)). Y, por supuesto, la opción óptima es instalar una distribución GNU/Linux directamente en el ordenador (por ejemplo, se puede recomendar la instalación de [Ubuntu](https://www.ubuntu.com/)), de modo que al arrancar el sistema se pueda elegir entre ambos sistemas operativos.


El sistema operativo de **Apple**, MacOSX, es de tip Unix, por tanto se cuenta directamente con la opción de usar una shell de Unix en una terminal (ve a Aplicaciones --> Utilidades
--> Terminal para abrir una terminal).

3.- Primeros pasos con la shell Unix
--------------------------------

La shell Unix suele utilizarse dentro de ventanas de texto conocidas como *consolas* o *terminales* de texto.
Cuando se abre una terminal, ésta muestra un indicador, que es conocido como **prompt**, para indicar que está esperando una entrada. En estas notas se usará generalmete un signo *\$* para indicar un prompt de Unix, aunque tu sistema puede tener un prompt algo diferente. Por ejemplo, el nombre del usuario y el nombre del ordenador suelen aparecer como información específica en el prompt, con frecuencia antecediendo a un signo *\$*. 

Habitualmente, el prompt muestra también el nombre de un directorio (o carpeta) que será aquella en la que realices operaciones con ficheros (por ejemplo, copiar o borrar ficheros). Este directorio se llama directorio de trabajo o directorio actual. La primera vez que se habre la terminal, el directorio de trabajo se situará en una carpeta de inicio que es específica a cada usuario. Por ejemplo, para el usuario *juandelacosa* su carpeta de inicio será


    /home/juandelacosa

Las órdenes *pwd*, *ls* y *cd*
----------

El nombre del la orden *pwd* proviene de la expresión "print working directory" (imprimir directorio de trabajo) y sirve para indicar la ruta completa del directorio donde se está trabajando actualmente. Por  ejemplo, cuando el usuario *juandelacosa* abre su terminal podría ejecutar *pwd*, obteniendo el siguiente resultado:

    $ pwd
    /home/juandelacosa

Para cambiar a otro directorio de trabajo, se usa el comando *cd*, que viene de "change directory" (cambiar directorio). Por ejemplo para subir al directorio situado un nivel superior, el usuario *juandelacosa* podría escribir la ruta absoluta  (comenzando con "/") iguiente:
    $ cd /home
    $ pwd
    /home
    
Para subir un nivel, también se puede utilizar la expresión ".." como en el siguiente ejemplo:
    $ cd ..
    $ pwd
    /

La última línea "/" designa el *directorio raíz* del sistema de ficheros. A partir de él, cuelga todo el sistema de carpetas y ficheros del sistema.

    Para volver a su carpeta de inicio, el usuario podría escribir la ruta absoluta:
    $ cd /home/juandelacosa
    $ pwd
    /home/juandelacosa

O bien podría escribir simplemente

    $ cd ~
        
O incluso

    $ cd


ls
--

*ls* (del inglés *list*) se usa para listar los ficheros y directorios contenidos en directorio actual de trabajo. Como muchas órdemes, *ls* puede tomar *argumentos* y *opciones*. Las **opciones** se utilizan para indicar a la shell información adicional sobre la acción de la orden, y suelen ir precedido de usn guión. Por ejemplo:

    $ ls -l

La opción *-l* indica a *ls* que proporcione un listdo largo, que contendrá información adicional sobre cada archivo (como que tamaño tiene, quién es el dueño, cuando se modificó por última vez, etc.). Por ejemplo, los primeros caracteres nos indican qué usarios tienen permiso para leer, escribir o ejecutar el archivo, ver
[\[Wikipedia\]](http://en.wikipedia.org/wiki/File_system_permissions).

Las órdenes, en ocasiones, también toman **argumentos** sobre los que se aplicará la opción de la orden. En el caso de *ls*, puedes especificar una lista de archivos para
aplicarles *ls*. Por ejemplo, si sólo queremos listar la información
sobre un archivo específico llamado "mifichero":

    $ ls -l mifichero

Se puede usar el carácter *comodín* **\*** para relacionar más de un
archivo:

     $ ls *.x

Si se teclea

    $ ls -F

entonces los directorios se mostrarán con una / y los archivos
ejecutables con un asterisco, lo que es útil para distinguirlos de archivos
ordinarios.

Cuando se se escribe *ls* sin argumentos, en general, se muestra la mayoría de los
archivos y subdirectorios. Pero ¡no todos!, en general. Por defecto
no se muestran los archivos y directorios que comiencen con un punto. Estos son archivos "ocultos". Entre ellos están "." (directorio actual) y ".." (directorio superior). Como hemos visto,  El último se usa frecuentemente para moverse un nivel hacia arriba en la estructura de directorios via:

    $ cd ..

Para listar todos los archivos ocultos se puede usar la opción -a:

    $ ls -aF

En esta orden hemos combinado las opciones -a y -F en una sola, -aF. Esto se puede hacer para todas las opciones, en genral.

Para obtener más infomación sobre la orden *ls*, se puede escribir:

    $ man ls

Esta orden ejecuta el comando *man* (páginas del manual) con el
argumento *ls*, y provoca que Unix muestre la información del manual disponible sobre *ls*.

Si haces esto, probablemente obtengas una página de información con ':'
en la parte inferior. En este punto estas en un entorno de la shell diferente, diseñado para permitirte desplazarte o buscar a través de un gran archivo. Los ':' son el prompt de este nuevo entorno shell. Los
comandos que puedes escribir en este punto son diferentes de los de la
shell de Unix. Los más útiles son:

    : q [para salir de esta shell y volver a Unix]
    : <SPACE> [para mostrar la siguiente pantalla completa]
    : b [vuelve a la pantalla completa anterior]

Editores de texto. Ordenes more, less, cat, head, tail
---------------------------

Existen dos excelentes editores de texto para la terminal Unix llamados *vim* y *emacs* (no preguntes cuál de ellos es mejor, excepto si te gustan las guerras santas). Si te interesa usar la terminal en el futuro, es recomendable buscar información sobre ellos y aprender a usarlos. Además, está disponible un editor muy sencillo (y mucho menos potente) llamado *nano*. Para editar un fichero basta escribir:

    $ nano mifichero
    
Para grabar se escribe `^O` y para salir `^X` (esta información se ofrece dentro del editor), donde `^` denota a la tecla `Ctrl`.

Las técnicas para avanzar páginas que fueron antes descritas al hablar de la orden *man* pueden aplicarse para ver el contenido archivos largor Paera ello, se puede usand el comando *less* (o bien el comando original de Unix *more*, algo menos sofisticado).

    $ less mi_archivo

Esta orden mostrará la primera pantalla completa del archivo llamado *mi_archivo* y mostrará el prompt ":".

El comando *cat* imprime el archivo completo en lugar de una página cada vez. *cat* proviene de "*cat*enate" (concatenar).  También puede usarse para mostrar el contenido de multiples archivos, consecutivamente, por ejemplo

    $ cat file1 file2 file3

El contenido de estos tres archivos se mostrará en pantalla, en el orden dado.

A veces, solo quieres ver las primeras 10 líneas o las últimas 5 líneas del mismo, por ejemplo. Prueba:

    $ head -10 filename
    
    $ tail -5 filename
    

Eliminando, moviendo, copiando archivos
---------------------------------------

Si quieres deshacerte de un archivo llamado *mibasura*, puedes usar la orden *rm*:

    $ rm mibasura
    
Si prefieres que el sistema te pida confirmación antes de borrar, puedes usar la opción -i: 

    $ rm -i mibasura
    rm: ¿borrar el fichero 'mibasura'? (s/n)

Se puede usar la orden *alias* para hacer esto por defecto:

    $ alias rm='rm -i'

El problema es que, al cerrar la terminal, el alias desaparecerá. La solución es incluir la siguiente línea en el fichero oculto `.bashrc` (situado en la carpeta de inicio del usuario):

    alias rm='rm -i'

Si quieres forzar el borrar sin preguntar (esto es útil si estás eliminando un grupo de archivos de una vez y si estás seguro de lo que estás haciendo), usa la opción -f. Por ejemplo, la siguiente orden borrará todos los ficheros con la extensión *.jpg* de la carpeta actual:
 
     $ rm -f *.jpg

La orden rm no borra directorios, excepto si usas la opción -r  (borrado recursivo, cuidado, ¡¡escríbelo solamente si estás seguro de lo que haces¡¡)
 
Para renombrar un archivo o moverlo a un lugar diferente (por ejemplo, un directorio diferente):

    $ mv oldfile newfile

puede ser una ruta completa o relativa a una localización fuera del
directorio actual de trabajo.

Para copiar un archivo usa *cp*:

    $ cp oldfile newfile

El original *oldfile* todavía existe. Para copiar un estructura entera recursivamente (copiando todos los archivos y cualquier directorio de la
misma manera), usa "cp -r":

    $ cp -r olddir newdir

Redirección de la entrada/salida
---------------------------------------------

En ocasiones, nos puede interesar que la salida de una orden se  dirija hacia un fichero, en vez de hacia la pantalla (que en principio es la salida estándar). Para ello se utiliza el carácter"&gt;" y a continuación el nombre del fichero. Por ejemplo:

    $ cat fich1 fich2 fich3 > ficherogrande
    
En este caso, el contenido de los ficheros `fich1` `fich2` `fich3` se  escribirá en`ficherogrande`. Atención, ¡el contenido previo de *ficherogrande* será eliminado!

Si deseamos conservar el contenido del último fichero podemos usar los caracteres "&gt;&gt;". Por ejemplo, la orden 

    $ cat fich1 fich2 fich3 >> ficherogrande
    
añadirá el contenido de los tres ficheros al final del último.

En ocasiones, deseamos que la salida de una orden se redirija hacia otra orden. Por ejemplo, consideremos la orden **sort**, que puede utilizarse para ordenar alfabéticamente las lineas de un fichero:

    $ cat mifichero
    Hola
    Adios
    Fin
    
    $ sort mifichero 
    Adios
    Fin
    Hola
    
Si deseamos ordenar alfabéticamente las 17 últimas lineas de un fichero llamado "listapalabras" podemos escribir:

    $ tail -17 | sort
    
El carácter "|" establece una "tubería" (en inglés, *pipe*) redirigiendo la salida de *tail -17* hacia la entrada de la orden sort.


Otro ejemplo: la orden **wc** cuenta el número de líneas, palabras y caracteres de un fichero. Las opciones -l, -w y -c devuelven respectiamente las líneas, palabras y caracters. Por ejemplo, sabiendo que el fichero `/usr/share/dict/spanish` está formado por líneas que contienen las palabras de un diccionario español, podemos escribir

    $ wc -w /usr/share/dict/spanish 
    86016 /usr/share/dict/spanish
    
Es decir, el diccionario tiene 86016 palabras. También podríamos haber usado *grep* de la siguiente forma:

    $ cat /usr/share/dict/spanish | wc -l
    86016


La orden **grep** se utiliza para buscar las conicidencias una palabro o una expresión en un fichero. Por ejemplo, la orden *grep* proporcionará el siguiente resultado:

    $ grep pago /usr/share/dict/spanish
    areópago
    campago
    ectópago
    galápago
    impago
    pago
    pagoda
    pagote
    relámpago
    
La orden anterior sería equivalente a

    $ cat /usr/share/dict/spanish | grep pago
     
Podríamos incluso encadenar varias tuberías para contar las palabras en español que contienen la expresión "pago":

    $cat /usr/share/dict/spanish  | grep pago | wc -w
    9

O cualquier otra expresión `;-)`

    $ cat /usr/share/dict/spanish  | grep rafa | wc -w
    58


Trabajos en primer y segundo plano
----------------------------------

Cuando ejecutas un programa que tardará mucho en ejecutarse, quizás
quieras ejecutarlo en *segundo plano* y así puedes seguir usando la
línea de comando de Unix para hacer otras cosas mientras se ejecuta. Por
ejemplo, supongamos que *fortrancode.exe* es un ejectutable de Fortran
en tu directorio actual que va a ejecutarse durante mucho tiempo. Puedes
hacer:

    $ ./fortrancode.exe &
    [1] 15442

si ahora presionas return deberías poder volver al prompt de Unix y
poder continuar trabajando.

./ antes del comando en el ejemplo superior es para decirle a Unix que
ejecute el ejecutable en este directorio (ver paths), y el & al final de
la línea le dice que lo haga en segundo plano. El "\[1\] 15442"
significa que es el trabajo en segundo plano número 1 ejectutado por
esta shell y que tiene el *processor id* 15442.

Si quieres encontrar que trabajos estás ejecutando en segundo plano y
sus pid's, prueba::..

> \$ jobs -l \[1\]+ 15443 Running ./fortrancode.exe &

Puedes traer un trabajo de vuelta al primer plano con:

    $ fg %1

Ahora no obtendrás el prompt de Unix hasta que el trabajo (o hasta que
lo mandes de vuelta al segundo plano como se describe abajo). El %1 se
refiere al trabajo 1. En este ejemplo *fg* solo hubiera bastado ya que
sólo hay un trabajo ejecutándose, pero en general tendrás varios en
segundo plano.

Para poner un trabajo que está en primer plano en segundo plano, puedes
escribir &lt;ctrl&gt;-z, que pausará el trabajo y te devolverá el
prompt:

    ^Z
    [1]+  Stopped                 ./fortrancode.exe
    $ 

Nota que el trabajo no está corriendo en segundo plano ahora, se ha
parado. Para hacer que se ejecute de nuevo en segundo plano, escribe:

    $ bg %1

O podrías mantenerlo ejecutándose en primer plano con "fg %1".

nice y top
----------

Si estás ejecutando un código que se ejecutará durante mucho tiempo
quizás quieras asegurarte que no enlentece otras cosas que estés
haciendo. Puedes hacer esto con el comando *nice*, por ejemplo:

    $ nice -n 19 ./fortrancode.exe &

le da al trabajo la menor prioridad (en nice puedes usar valores entre 1
y 19) así no acapara la CPU si estás intentando editar un archivo al
mismo tiempo, por ejemplo.

You can change la prioridad de un trabajo ejecutándose en segundo plano
con *renice*, por ejemplo:

    $ renice -n 19 15443

donde el último número es la identidad del proceso.

Otro comando útil es *top*. Esto llenará tu ventana con una página de
información sobre los trabajos que se están ejecutando en tu ordenador
que usan la mayor parte de los recursos. Ver topcommand para algunos
ejemplos.

matando trabajos
----------------

A veces necesitas matar un trabajo que se está ejecutando, quizás porque
te has dado cuenta de que se va a estar ejecutando durante mucho tiempo,
o porque le diste unos datos erróneos de entrada. O porque estás
ejecutando un programa como la shell de IPython, se congela y no hay
manera de volver a coger el control. (Esto a veces ocurre cuando dibujas
con el comando *pylab.show()*, por ejemplo).

Se puede acabar con muchos programas con &lt;ctrl&gt;-c. Para que esto
funcione, el trabajo debe estar ejecutándose en primer plano, así que
quizás necesites usar el comando *fg*.

A veces, esto no funciona, como cuando se congela IPython. Entonces
intenta pararlo con &lt;ctrl&gt;-z (que debería funcionar), descubre el PID, y usa el comando *kill*.

    $ jobs -l \[1\]
    15841 Suspended ipython
    $ kill 15841

Pulsa enter de nuevo y con suerte verás:

    $
    [1]+ Terminated              ipython
    $ 

Si no, es necesario una acción más drástica con la señal -9:

    $ kill -9 15841

archivo .bashrc
---------------

Cada vez que inicias una nueva shell bash, por ejemplo, con el comando
anterior, cuando accedes por primera vez o cuando abres una nueva
ventana (asumiendo que por defecto es en bash), un archivo llamado
".bashrc" de tu directorio home se ejecuta como un script de bash.
Puedes colocar en este archivo cualquier cosa que quieras que se ejecute
cuando se inicie, como exportar variables de entorno, establecer las
rutas, definir alias, establecer el prompt de la manera que quieras,
etc. Mira abajo para más cosas.

Variables de entorno
--------------------

EL comando *printenv* mostrará todas las variables de entorno
establecidas, por ejemplo:

    $ printenv
    USER=rjl
    HOME=/Users/rjl
    PWD=/Users/rjl/uwhpsc/sphinx
    FC=gfortran
    PYTHONPATH=/Users/rjl/claw4/trunk/python:/Applications/visit1.11.2/src/lib:
    PATH=/opt/local/bin:/opt/local/sbin:/Users/rjl/bin
    etc.

También puedes mostrar una sola variable, por ejemplo:

    $ printenv HOME
    /Users/rjl

o:

    $ echo $HOME
    /Users/rjl

La último forma tiene \$HOME en lugar de HOME porque een realidad
estamos *usando* la variable en un comando echo en lugar de sólamente
imprimiendo su valor. Essta variable particular es útil para un montón
de cosas como

> \$ cd \$HOME/uwhpsc

que irá al subdirectorio uwhpsh de tu directorio home no importa donde
empieces.

Como parte de la Tarea 1 se te pide definir una nueva variable de
entorno para hacer esto incluso más sencillo, por ejemplo mediante:

    $ export UWHPSC=$HOME/uwhpsc

Date cuenta que no hay espacios alredodor de =. Esto define una nueva
variable y la *exporta*, así que puede ser usada por otros programas que
ejecutes en esta shell (esto no es muy importante para nuestro
propósitos, pero a veces es necesario).

Ahora puedes hacer sólo esto:

    $ cd $UWHPSC

para ir a este directorio.

Nota que he definido una variable de entorno FC como:

    $ printenv FC
    gfortran

Esta variable de entorno se usa en algunos Makefiles (ver makefiles)
para determinar que compilador de Fortran se usará al compilar códigos
en Fortran.

PATH y otras rutas de búsqueda
------------------------------

Cuandp escribas un comando en el prompt de Unix, la shell busca un
programa para ejecutar. Esto es cierto para los comandos incorporados y
también para los nuevos comandos que definas o para los programas que
hallas instalados. Para saber donde buscar dichos programas, la shell
busca en los directorios especificados por la variable PATH (ver env
arriba). Esto puede ser algo como:

    $ printenv PATH
    PATH=/usr/local/bin:/usr/bin:/Users/rjl/bin

Esto da una lista de directorios de búsqueda, en orden, separadas por
":". PATH es en general más largo que esto, pero en el ejemplo superior
hay 3 directorios en la ruta. Los primeros dos son reposiorios generales
de todo el sistema y el último es mi propio *bin* directory (bin viene
de binario ya que los ejecutables son frecuentemente archivos binarios,
además a menudo los directorios bin también contiene scripts de shell u
otros programas en texto).

which
-----

El comando *which* es útil para averiguar la ruta completa del código
que está siendo actualmente ejecutado cuando escribes un comando, por
ejemplo:

    $ which gfortran
    /usr/bin/gfortran

    $ which f77
    $

En el último caso no se encuentra ningún programa llamado f77 en la ruta
de búsqueda, o porque no está instalado o porque el directorio no está
en PATH.

Algunos programas requieren que se establezca su propia ruta si
necesitan buscar archivos de entrada. Por ejemplo, puedes establecer
MATLABPATH o PYTHONPATH para que sea una lista de directorios (separados
por ":") para buscar archivos .m cuando ejecutes un comando en Matlab o
archivos .py cuando importes módulos en Python,
